**Funcion `findAll`**



```python
def findAll(S,T):
  """
  It receives two parameters:
    S - the whole String where T needs to be found
    T - the target String to be found

  It returns a list L where:
    L[0] - N = number of times T was found inside of S
    L[1] - (initPos, finalPos) of the 1st occurrence of T
    ...
    L[n-1] - (initPos, finalPos) of the n-th occurrence of T
  """
  n = len(S)    # n = length of S
  m = len(T)    # m = length of T

  # There are three special cases which don't require iterations to solve:

  # If T is longer than S, T can't be inside S   
  if m > n:                  
    return [0]
  # If both strings have the same length, there are only two options: 
  if m == n:                 
    if T == S:                   
      return [1, (0,len(S)-1)]    # If T = S, T is inside S 
    elif T == S:                       
      return [0]                  # If T != S, T is not inside S 

  # For the rest of cases, an iteration process must be followed:

  #The default list starts with 0 occurrences
  resultList = [0]
  incidencesCounter = 0
  #This loop iterates through S, from index 0 until the last index where T might start
  for i in range(n - m + 1):
    word = ""
    # If the first character of T is a match, it gets added to the potential word
    if S[i] == T[0]:
      word += T[0]
      #The next m - 1 characters of S get compared one by one with the characters of T
      for j in range(1, m):
        if S[i + j] == T[j]:    
          word += T[j]    #If the character matches, it gets added to the potential word
        else:
          break           #If even a single character doesn't match, the process gets aborted    
      # Once a new occurrence is found, the counter gets incremented
      # and the (initialPosition, finalPosition) tuple is added to the list
      if word == T:
        incidencesCounter += 1
        resultList.append((i, i + m - 1))
  #Finally, the number of matches is updated
  resultList[0]= incidencesCounter    
  return resultList                   
```

**Hallando el segmento asignado**

```python
import time

segment = "GTTATCTTACTTCTTCTTCTAAAACA"
print("Genome segment to be found:", segment)

inTime = time.time()
result = findAll(secuence, segment)   # Using the function to find the given sequence
endTime = time.time()

print("Result:", result, end = "\n\n")
quantity = result[0]

# Let's give a nice output format to the results
if quantity == 1:
  print("This means that there is 1 occurrence of this segment in the COVID genome.\n")
else:
  print("This means that there are", result[0], "occurrences of this segment in the COVID genome.\n")

if quantity != 0:
  for i in range(1, quantity + 1):
    print("The occurrence #" + str(i), "starts at index", result[i][0], "and ends at index", str(result[i][1]) + ".")

print("\nTime elapsed:", round(endTime-inTime, 6), "seconds.")
```
